```python
import json

S1_JSON_PATH = "/content/drive/MyDrive/sen12_dataset/SEN12FLOOD-dataset/SEN12FLOOD/S1list.json"

with open(S1_JSON_PATH, "r") as f:
    s1_data = json.load(f)

valid_event_id = None

for event_id, event in s1_data.items():
    pre, flood = [], []
    for k, v in event.items():
        if k.isdigit():
            if v["FLOODING"]:
                flood.append(v["filename"])
            else:
                pre.append(v["filename"])
    if len(pre) > 0 and len(flood) > 0:
        valid_event_id = event_id
        break

print("Selected flood event:", valid_event_id)
print("Pre-flood count:", len(pre))
print("Flood count:", len(flood))
```

```python
pre_fname = pre[0]
flood_fname = flood[0]

print("Pre-flood base name:", pre_fname)
print("Flood base name:", flood_fname)
```

```python
import os

EVENT_FOLDER = f"/content/drive/MyDrive/sen12_dataset/SEN12FLOOD-dataset/SEN12FLOOD/0200"

def find_vv_vh(prefix):
    files = os.listdir(EVENT_FOLDER)
    vv = [f for f in files if f.startswith(prefix) and "_VV.tif" in f][0]
    vh = [f for f in files if f.startswith(prefix) and "_VH.tif" in f][0]
    return vv, vh

pre_vv, pre_vh = find_vv_vh(pre_fname)
flood_vv, flood_vh = find_vv_vh(flood_fname)

print("Pre VV:", pre_vv)
print("Pre VH:", pre_vh)
print("Flood VV:", flood_vv)
print("Flood VH:", flood_vh)
```

```python
import rasterio
import numpy as np

def load_tif(path):
    with rasterio.open(path) as src:
        return src.read(1).astype(np.float32)

pre_vv_arr = load_tif(os.path.join(EVENT_FOLDER, pre_vv))
pre_vh_arr = load_tif(os.path.join(EVENT_FOLDER, pre_vh))
flood_vv_arr = load_tif(os.path.join(EVENT_FOLDER, flood_vv))
flood_vh_arr = load_tif(os.path.join(EVENT_FOLDER, flood_vh))

print(pre_vv_arr.shape, flood_vv_arr.shape)
```

```python
def center_crop(arr, target_h, target_w):
    h, w = arr.shape
    start_h = (h - target_h) // 2
    start_w = (w - target_w) // 2
    return arr[start_h:start_h+target_h, start_w:start_w+target_w]
```

```python
heights = [
    pre_vv_arr.shape[0],
    pre_vh_arr.shape[0],
    flood_vv_arr.shape[0],
    flood_vh_arr.shape[0]
]
widths = [
    pre_vv_arr.shape[1],
    pre_vh_arr.shape[1],
    flood_vv_arr.shape[1],
    flood_vh_arr.shape[1]
]
target_h = min(heights)
target_w = min(widths)

print("target size:" , target_h, target_w)
```

```python
pre_vv_arr = center_crop(pre_vv_arr, target_h, target_w)
pre_vh_arr = center_crop(pre_vh_arr, target_h, target_w)
flood_vv_arr = center_crop(flood_vv_arr, target_h, target_w)
flood_vh_arr = center_crop(flood_vh_arr, target_h, target_w)

print(pre_vv_arr.shape, flood_vv_arr.shape)
```

```python
X = np.stack([
    pre_vv_arr,
    pre_vh_arr,
    flood_vv_arr,
    flood_vh_arr
], axis = 0)

print(X.shape)
```

step 3 sar normalization

```python
def log_transform(arr, eps=1e-6):
  return np.log(arr+eps)
```

```python
X_log = log_transform(X)
```

```python
def normalize_per_channel (X):
  X_norm = np.zeros_like(X)
  for i in range(X.shape[0]):
    mean = X[i].mean()
    std = X[i].std() + 1e-6
    X_norm[i] = (X[i]-mean)/std
  return X_norm
```

```python
X_norm = normalize_per_channel(X_log)
```

```python
for c in range(4):
  print(
      f"channel: {c}: mean: {X_norm[c].mean():.2f}, std: {X_norm[c].std():.2f}"
  )
```
